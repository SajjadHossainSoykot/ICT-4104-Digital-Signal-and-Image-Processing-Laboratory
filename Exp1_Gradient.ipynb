# Experiment No. 1

## Title
**To create an OpenCV-Python program to get a gradient image from an RGB/Grayscale image**

## Objective
To generate the gradient image of a given grayscale image and observe intensity variation between neighboring pixels.

## Mathematical Theory
Gradient represents the rate of intensity change in an image. If pixel intensity changes rapidly, that region usually contains an edge.

For a horizontal discrete case,

$$
G(x,y) = |f(x,y+1) - f(x,y)|
$$

Where:

- $f(x,y)$ = input pixel intensity
- $G(x,y)$ = gradient value

Large gradient values indicate strong intensity transition.

## Instruments / Requirements
- Computer or Laptop
- Python
- OpenCV
- NumPy
- Input image: `Exp1_Demo.jpg`

## Procedure
1. Read the input image in grayscale mode.
2. Take two neighboring horizontal pixels.
3. Compute their absolute intensity difference.
4. Store the result in a new image.
5. Display the original and gradient images.

## Output
- Original grayscale image
- Gradient image showing intensity changes
- Output image: `Exp1_Output.jpg`

## Result
The gradient image was successfully generated. Edge regions became brighter because the difference between neighboring pixel values is higher in those regions.

## Conclusion
This experiment shows that gradient operation is useful for identifying edges and sudden intensity transitions in an image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image in grayscale mode
img = cv2.imread('demo_image.jpg', 0)
# img = cv2.imread('Exp1_Demo.jpg', 0)

# Get image size
h, w = img.shape

# Create empty output image
gradient = np.zeros((h, w), dtype=np.uint8)

# Apply gradient magnitude formula:
# Gx = f(x,y+1) - f(x,y)
# Gy = f(x+1,y) - f(x,y)
# G  = sqrt(Gx^2 + Gy^2)
for i in range(h - 1):
    for j in range(w - 1):
        gx = int(img[i, j + 1]) - int(img[i, j])
        gy = int(img[i + 1, j]) - int(img[i, j])

        g = np.sqrt(gx**2 + gy**2)

        if g > 255:
            g = 255

        gradient[i, j] = int(g)

# Display using matplotlib
plt.figure(figsize=(10, 8))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(gradient, cmap='gray')
plt.title("Gradient Image")
plt.axis("off")

plt.tight_layout()
plt.savefig('Exp1_Output.png')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image in grayscale mode
img = cv2.imread('demo_image.jpg', 0)
# img = cv2.imread('Exp1_Demo.jpg', 0)

# Get image size
h, w = img.shape

# Create empty output images
gx_img = np.zeros((h, w), dtype=np.uint8)
gy_img = np.zeros((h, w), dtype=np.uint8)
gradient = np.zeros((h, w), dtype=np.uint8)

# Apply gradient formulas:
# Gx = f(x,y+1) - f(x,y)
# Gy = f(x+1,y) - f(x,y)
# G  = sqrt(Gx^2 + Gy^2)

for i in range(h - 1):
    for j in range(w - 1):
        gx = int(img[i, j + 1]) - int(img[i, j])
        gy = int(img[i + 1, j]) - int(img[i, j])

        # For displaying X and Y gradient, use absolute value
        gx_abs = abs(gx)
        gy_abs = abs(gy)

        # Final gradient magnitude
        g = np.sqrt(gx**2 + gy**2)

        # Clip values to 0-255
        if gx_abs > 255:
            gx_abs = 255

        if gy_abs > 255:
            gy_abs = 255

        if g > 255:
            g = 255

        gx_img[i, j] = int(gx_abs)
        gy_img[i, j] = int(gy_abs)
        gradient[i, j] = int(g)

# Display using matplotlib
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(gx_img, cmap='gray')
plt.title("X Gradient Image")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(gy_img, cmap='gray')
plt.title("Y Gradient Image")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(gradient, cmap='gray')
plt.title("Final Gradient Magnitude Image")
plt.axis("off")

plt.tight_layout()
plt.savefig('Exp1_Output.png')
plt.show()
```
