# Experiment No. 2

## Title
**To create an OpenCV-Python program to show a histogram of an RGB/Grayscale image**

## Objective
To display and analyze the histogram of a grayscale image.

## Mathematical Theory
A histogram shows the frequency distribution of pixel intensities in an image. For a grayscale image, intensity values range from 0 to 255.

If $n_k$ is the number of pixels with intensity $r_k$, then:

$$
h(r_k) = n_k
$$

The normalized histogram is:

$$
p(r_k) = \frac{n_k}{MN}
$$

Where:

- $M \times N$ = total number of pixels
- $p(r_k)$ = probability of intensity $r_k$

Histogram helps analyze brightness and contrast distribution.

## Instruments / Requirements
- Computer or Laptop
- Python
- OpenCV
- NumPy
- Matplotlib
- Input image: `Exp2_Demo.jpg`

## Procedure
1. Read the image in grayscale mode.
2. Create an array for 256 intensity levels.
3. Count the occurrence of each pixel intensity.
4. Plot the frequency distribution.
5. Observe the intensity distribution of the image.

## Output
- Histogram of grayscale intensities from 0 to 255
- Output image: `Exp2_Output.jpg`

## Result
The histogram was successfully plotted and showed the distribution of pixel values in the image.

## Conclusion
Histogram is a useful tool for analyzing image brightness, darkness, and intensity spread.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image in grayscale
img = cv2.imread('Exp2_Demo.png', 0)

# Create histogram array
hist = np.zeros(256, dtype=int)

# Count intensity frequency manually
h, w = img.shape
for i in range(h):
    for j in range(w):
        intensity = img[i, j]
        hist[intensity] += 1

# Display original image and histogram
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.plot(hist, color='black')
plt.title("Histogram")
plt.xlabel("Intensity Value")
plt.ylabel("Frequency")
plt.xlim([0, 255])
plt.grid()

plt.tight_layout()
plt.savefig('Exp2_Output.png')  # Save full figure
plt.show()
```
