## Title
**To create an OpenCV-Python program to enhance brightness of a given RGB/Grayscale image**

## Objective
To increase the brightness of both grayscale and RGB images using a simple pixel-wise transformation.

## Mathematical Theory
Brightness enhancement is a point processing operation where a constant value is added to each pixel intensity.

For a grayscale image:

$$
g(x,y) = f(x,y) + b
$$

For an RGB image:

$$
g(x,y,c) = f(x,y,c) + b
$$

Where:

- $f(x,y)$ = original pixel intensity  
- $g(x,y)$ = output pixel intensity  
- $b$ = brightness constant  
- $c$ = color channel (Red, Green, Blue)

To maintain valid intensity range:

$$
g(x,y) = \min(f(x,y) + b,\ 255)
$$

## Instruments / Requirements
- Computer or Laptop  
- Python  
- OpenCV  
- NumPy  
- Matplotlib  
- Input image: `Exp5_Demo.jpg`

## Procedure
1. Read the image (grayscale or color) using OpenCV.
2. If the image is color, convert it from BGR to RGB format.
3. Define a brightness constant value.
4. Add the constant value to each pixel (and each channel for RGB).
5. Clip pixel values to the range 0–255.
6. Store the result in a new image.
7. Display the original and brightness-enhanced images.

## Output
- Original image (Grayscale/RGB)  
- Brightness enhanced image  

## Result
The brightness-enhanced image shows increased intensity across all pixels. In RGB images, all color channels are uniformly increased, resulting in a brighter and more visually clear image.

## Conclusion
Brightness enhancement is a fundamental image processing technique used to improve visibility in low-light images. It works effectively for both grayscale and RGB images by uniformly increasing pixel intensity values.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read color image
img = cv2.imread('demo_image.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Brightness constant
b = 50

h, w, c = img.shape
bright = np.zeros((h, w, c), dtype=np.uint8)

# Apply brightness enhancement manually
for i in range(h):
    for j in range(w):
        for k in range(c):  # RGB channels
            value = int(img[i, j, k]) + b
            if value > 255:
                value = 255
            bright[i, j, k] = value

# Display images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(bright)
plt.title("Brightness Enhanced Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
