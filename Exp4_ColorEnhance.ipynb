# Experiment No. 4

## Title
**To create an OpenCV-Python program to enhance color of a given RGB/Grayscale image**

## Objective
To enhance the color information of an RGB image by increasing saturation.

## Mathematical Theory
Color enhancement is commonly performed in HSV color space.

HSV means:
- $H$ = Hue
- $S$ = Saturation
- $V$ = Value

To increase color vividness, saturation is multiplied by a factor:

$$
S' = kS
$$

Where $k > 1$ increases saturation.

If the value exceeds 255, it is clipped:

$$
S' = \min(255, kS)
$$

## Instruments / Requirements
- Computer or Laptop
- Python
- OpenCV
- NumPy
- Input image: `Exp4_Demo.png`

## Procedure
1. Read the color image.
2. Convert it from BGR to HSV color space.
3. Increase the saturation channel.
4. Clip values above 255.
5. Convert the image back to BGR/RGB.
6. Display the original and enhanced images.

## Output
- Original color image
- Color enhanced image

## Result
The output image became more vivid because the saturation values were increased.

## Conclusion
This experiment shows that color enhancement can be performed effectively by increasing the saturation component in HSV color space.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read color image
img = cv2.imread('Exp4_Demo.png')

# Convert BGR to HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Split channels
H, S, V = cv2.split(hsv)

# Saturation multiplication factor
k = 1.5

# Increase saturation manually
h, w = S.shape
for i in range(h):
    for j in range(w):
        new_val = int(S[i, j] * k)
        if new_val > 255:
            new_val = 255
        S[i, j] = new_val

# Merge and convert back
enhanced_hsv = cv2.merge((H, S, V))
enhanced_bgr = cv2.cvtColor(enhanced_hsv, cv2.COLOR_HSV2BGR)

# Convert BGR to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
enhanced_rgb = cv2.cvtColor(enhanced_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(enhanced_rgb)
plt.title("Color Enhanced Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
